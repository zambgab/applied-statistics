```python
#Interquartil = Q3 – Q1
iqr = q3 - q1

#Limite 1 = Q3 + 3*IQ
limite1 = q3 + 3*iqr

#Limite 2 = Q3 + 1,5*IQ
limite2 = q3 + 1.5*iqr

#Limite 3 = Q1 – 1.5*IQ
limite3 = q1 - 1.5*iqr

#Limite 4 = Q1 – 3*IQ
limite4 = q1 - 3*iqr
```

### Cálculo do coeficiente de variação (CV).

#### Ele permite comparar a dispersão de diferentes conjuntos de dados, mesmo quando possuem unidades de medidas distintas.

##### Um CV baixo indica que os dados estão mais próximos da média, ou seja, são mais homogêneos. Já um CV alto indica maior dispersão dos dados em relação à média.

#### Calculo = desvio padrão / média.

```python
# Cálculo do coeficiente de variação (CV)
# CV = desvio padrão/média

desc = cadastro.describe().T

desc['CV'] = desc['std']/desc['mean']
```

```python
desc.round(2)
```

Renda presumida: O desvio padrão é 10.76 vezes o valor da média. Isso indica presença de outliers.


    Recomendado é até 0,3 ou 30%. Valores acima de 30% indicam presença de outliers. (Critério do IBGE)
