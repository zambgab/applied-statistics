```python
#Primeiro quartil
q1 = np.percentile(cadastro['RENDA_PRESUMIDA'], 25)
q2 = np.percentile(cadastro['RENDA_PRESUMIDA'], 50)
q3 = np.percentile(cadastro['RENDA_PRESUMIDA'], 75)
```

```python
#Interquartil = Q3 ‚Äì Q1
iqr = q3 - q1

#Limite 1 = Q3 + 3*IQ
limite1 = q3 + 3*iqr

#Limite 2 = Q3 + 1,5*IQ
limite2 = q3 + 1.5*iqr

#Limite 3 = Q1 ‚Äì 1.5*IQ
limite3 = q1 - 1.5*iqr

#Limite 4 = Q1 ‚Äì 3*IQ
limite4 = q1 - 3*iqr
```

```python
print(round(iqr))
print(round(limite1))
print(round(limite2))
print(round(limite3))
print(round(limite4))
```

### C√°lculo do coeficiente de varia√ß√£o (CV).

#### Ele permite comparar a dispers√£o de diferentes conjuntos de dados, mesmo quando possuem unidades de medidas distintas.

##### Um CV baixo indica que os dados est√£o mais pr√≥ximos da m√©dia, ou seja, s√£o mais homog√™neos. J√° um CV alto indica maior dispers√£o dos dados em rela√ß√£o √† m√©dia.

#### Calculo = desvio padr√£o / m√©dia.

```python
# C√°lculo do coeficiente de varia√ß√£o (CV)
# CV = desvio padr√£o/m√©dia

desc = cadastro.describe().T

desc['CV'] = desc['std']/desc['mean']
```

```python
desc.round(2)
```

Renda presumida: O desvio padr√£o √© 10.76 vezes o valor da m√©dia. Isso indica presen√ßa de outliers.
